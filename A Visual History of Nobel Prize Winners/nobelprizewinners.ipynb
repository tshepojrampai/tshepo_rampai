## 3. USA dominance
<p>Not so surprising perhaps: the most common Nobel laureate between 1901 and 2016 was a man born in the United States of America. But in 1901 all the winners were European. When did the USA start to dominate the Nobel Prize charts?</p>
    

```python
# Proportion of USA born winners per decade
nobel['usa_born_winner'] = nobel["birth_country"]== "United States of America"
nobel["decade"]= (np.floor(nobel["year"]/10)*10).astype(int)

prop_usa_winners = (nobel.groupby("decade", as_index=False)["usa_born_winner"].mean())

# Proportion Display
prop_usa_winners
```


* The US didn't just dominate the Nobel Prize charts from the beggining:

     * It only had just above 7% in the next 2 decades.
     * In the 1930s decade, 25% of the Nobel Prize winners were from the USA. This is when they started their dominance because they have never reached a proportion lesser than that s

## 5. What is the gender of a typical Nobel Prize winner?
<p>So the USA became the dominating winner of the Nobel Prize first in the 1930s and had kept the leading position ever since.
```python
# Calculating the proportion of female laureates per decade
nobel['female_winner'] = nobel["sex"]== "Female"

prop_female_winners = nobel.groupby(["decade", "category"], as_index=False)["female_winner"].mean()
prop_female_winners

# Plot proportion of female winners by decade, grouped by prize category (physics, medicine, etc)
print("5.")
graph1 =sns.lineplot(x="decade",y="female_winner",hue= "category",data= prop_female_winners)
```

* On average, women get less prizes in Physics,  Economics, Chemistry and Medicine (Ranked from least to greatest).

## 8. How old are you when you get the prize?
<p>The list of repeat winners contains some illustrious names! We again meet Marie Curie, who got the prize in physics for discovering radiation and in chemistry for isolating radium and polonium. John Bardeen got it twice in physics for transistors and superconductivity, Frederick Sanger got it twice in chemistry, and Linus Carl Pauling got it first in chemistry and later in peace for his work in promoting nuclear disarmament. We also learn that organizations also get the prize as both the Red Cross and the UNHCR have gotten it twice.</p>
<p>But how old are you generally when you get the prize?</p>

```python
# Converting birth_date from String to datetime
nobel['birth_date'] = pd.to_datetime(nobel["birth_date"])
print("8.")


# Calculating the age of Nobel Prize winners
nobel['age'] = nobel["year"]-nobel["birth_date"].dt.year

# What is the mean age of winners and it's standard devation? What is the median age?
print("The mean is:")


print((nobel["age"]).mean())
print(" ")
print("The standard deviation is:")
print(np.std(nobel["age"]))
print(" ")
print("The median is:")
print((nobel["age"].median()))
```

* On average, a person can expect to win the Nobel prize between age 59 and 60 years.
* The mean of this data gives us this Expected Value.
* With a lower standard deviation, it is safe to say that the data is not spread out from the average(mean) on both sides of the mean.
* That is, the data is closely clusttered around the mean on average, because this data follows  normal distribution.
